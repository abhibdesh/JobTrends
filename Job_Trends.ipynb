# Dropping off the column which is to be predicted

# Label encoding to convert String Categorical variables in numerical

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
for i in categoricals_cols:
    placements[i]=le.fit_transform(placements[i])
```

# Scaling the percentages

```python
from sklearn.preprocessing import StandardScaler
scaling=["ssc_p","hsc_p","degree_p","etest_p","mba_p"]
sc = StandardScaler()
scaled=sc.fit_transform(placements[['ssc_p', 'hsc_p', 'degree_p', 'etest_p', 'mba_p']])
scaled = pd.DataFrame(scaled,columns=scaling)
placements=placements.drop(columns=scaling,axis=1)
placements=placements.merge(scaled, left_index=True, right_index=True, how = "right")
```

```python
placements.head()
```

# Predicting  if a student gets placed or not

# Separating dependent and in dependent variable

```python
X=placements[['gender','ssc_b','hsc_b','hsc_s','degree_t','workex','specialisation','ssc_p','hsc_p','degree_p','etest_p','mba_p']]
y=placements['status']
```

# Importing train_test_split

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=101)
```

# Prediction with Decision tree

```python
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier()
clf.fit(X_train,y_train)
```

```python
accuracy4 = clf.score(X_test,y_test)
print("Accuracy: {:.2f} %".format(accuracy4.mean()*100))
print("Standard Deviation: {:.2f} %".format(accuracy4.std()*100))
```

# Prediction with Gaussian Naive-Bayes 

```python
from sklearn.naive_bayes import GaussianNB
nb=GaussianNB()
nb.fit(X_train,y_train)
```

```python
accuracy5 = nb.score(X_test,y_test)
print("Accuracy: {:.2f} %".format(accuracy5.mean()*100))
print("Standard Deviation: {:.2f} %".format(accuracy5.std()*100))
```

# Prediction with Gradient Boosting Classifier

```python
from sklearn.ensemble import GradientBoostingClassifier
gbc=GradientBoostingClassifier()
gbc.fit(X_train,y_train)
```

```python
accuracy6 = gbc.score(X_test,y_test)
print("Accuracy: {:.2f} %".format(accuracy6.mean()*100))
print("Standard Deviation: {:.2f} %".format(accuracy6.std()*100))
```

# Salary Prediction

# Splitting the dataset

```python
A=placements[['workex','specialisation','ssc_p','hsc_p','degree_p','etest_p','mba_p']]
b=placements['salary']
from sklearn.model_selection import train_test_split
A_train,A_test,b_train,b_test=train_test_split(A,b,test_size=0.3,random_state=101)
```

# Prediction using Decision Tree Classifier

```python
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
clf2 = DecisionTreeClassifier()
clf2.fit(A_train,b_train)
accuracyZ = clf2.score(A_test,b_test)
print("Accuracy: {:.2f} %".format(accuracyZ.mean()*100))
print("Standard Deviation: {:.2f} %".format(accuracyZ.std()*100))
```

# Prediction using Gaussian Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB
nb2=GaussianNB()
nb2.fit(A_train,b_train)
accuracyY = nb2.score(A_test,b_test)
print("Accuracy: {:.2f} %".format(accuracyY.mean()*100))
print("Standard Deviation: {:.2f} %".format(accuracyY.std()*100))
```

 # Prediction using Gradient Boosting Classifier

```python
from sklearn.ensemble import GradientBoostingClassifier
gbc2=GradientBoostingClassifier()
gbc2.fit(A_train,b_train)
accuracyX = gbc2.score(A_test,b_test)
print("Accuracy: {:.2f} %".format(accuracyX.mean()*100))
print("Standard Deviation: {:.2f} %".format(accuracyX.std()*100))
```

# Prediction using Support Vector Machine (SVM) 

```python
import sklearn.svm as svm
svm = svm.LinearSVC()
svm.fit(A_train, b_train)
accuracyT = svm.score(A_test,b_test)
print("Accuracy: {:.2f} %".format(accuracyT.mean()*100))
print("Standard Deviation: {:.2f} %".format(accuracyT.std()*100))
```

# Prediction using K-Neighbors Classifier

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=10)
knn.fit(A_train, b_train)
accuracyR = knn.score(A_test,b_test)
print("Accuracy: {:.2f} %".format(accuracyR.mean()*100))
print("Standard Deviation: {:.2f} %".format(accuracyR.std()*100))
```

 K-Neighbors Classifier is giving the highest accuracy for predction of the salaries.
